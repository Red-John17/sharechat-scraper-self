```python
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
```

```python
# Function to scroll down to the comment button and click it
def scroll_to_comment_button(driver):
    try:
        comment_button = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'button[data-cy="comment-action-btn"]'))
        )
        driver.execute_script("arguments[0].scrollIntoView(true);", comment_button)
        comment_button.click()
    except Exception as e:
        print("An error occurred while scrolling to the comment button:", str(e))

# Function to extract data from elements with class name "Pt($md)"
def extract_data_from_elements(driver):
    try:
        author_elements = driver.find_elements(By.CSS_SELECTOR, 'div.Pt($md) strong[data-cy="author-name"]')
        for author_element in author_elements:
            author_name = author_element.text
            print("Like Name:", author_name)
    except Exception as e:
        print("An error occurred while extracting data:", str(e))
```

```python
# Function to locate and click on the specified span element by text content
def click_on_span_element(driver, span_text):
    try:
        span_element = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.XPATH, f'//span[text()="{span_text}"]'))
        )
        span_element.click()
    except Exception as e:
        print(f"An error occurred while clicking on the span element '{span_text}':", str(e))
```

```python
# URL of the ShareChat page
url = "https://sharechat.com/tag/ayJbb5?referer=tagProfileSearchPage"

# Initialize the Firefox driver
driver = webdriver.Firefox()

# Open the URL in Firefox
driver.get(url)

# Wait for the page to load (you can increase the wait time as needed)
time.sleep(10)

try:
    # Locate the elements containing the information
    author_element = driver.find_element(By.CSS_SELECTOR, 'strong[data-cy="author-name"]')

    parent_div = WebDriverWait(driver, 10).until(
        EC.visibility_of_element_located((By.XPATH, '//div[contains(@class, "D(f)") and not(contains(@class, " "))]'))
    )

    # Locate the child div elements within the parent div
    child_divs = parent_div.find_elements(By.TAG_NAME, "div")

    caption = driver.find_element(By.CSS_SELECTOR, 'div[data-cy="post-caption"]')

    # Wait for the parent div to be present
    post_div = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//div[contains(concat(" ", normalize-space(@class), " "), " Pos(r) Ov(h) ")]')))

   # Find the second div element within the parent div
    second_div = post_div.find_elements(By.TAG_NAME, "div")[1]

    img_element = second_div.find_element(By.TAG_NAME, "img")

    # Extract the src attribute value
    src_value = img_element.get_attribute("src")


    # Extract the text from the child div elements
    number_of_views = child_divs[0].text
    years_before = child_divs[2].text
    # Extract the text from the elements
    author_name = author_element.text
    caption_text = caption.text

    # time.sleep(8)

    # scroll_to_comment_button(driver)

    # time.sleep(8)

    # click_on_span_element(driver)

    # time.sleep(5)

    # extract_data_from_elements(driver)



    # Print the extracted data
    print("Author Name:", author_name)
    print("Number of Views:", number_of_views)
    print("Years Before:", years_before)
    print("Caption:", caption_text)
    print("Data:" , src_value)

except Exception as e:
    print("An error occurred:", str(e))

finally:
    # Close the WebDriver
    driver.quit()
```
